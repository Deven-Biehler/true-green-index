```python
import torch
from torch.utils.data import Dataset, DataLoader, random_split
from torchvision import transforms, models
import torch.nn as nn
import numpy as np
import tifffile as tiff
import os
from PIL import Image
import matplotlib.pyplot as plt
from collections import Counter

workspace_path = "/content/drive/MyDrive/Computer Vision/Final Project Notebook"
os.chdir(workspace_path)


config = {
    'image_paths': ['Seattle_Parks/sentinel_data.tiff',
                    # 'Sacremento_Parks/sentinel_data.tiff'
                    ],
    'mask_paths': ['Seattle_Parks/park_mask.tiff',
                  #  'Sacremento_Parks/park_mask.tiff'
                   ],
    'patch_size': 256,
    'batch_size': 16,
    'epoch': 100
}
```

```python
from torchvision import transforms

transform = transforms.Compose([
    # transforms.RandomHorizontalFlip(p=0.5),  # Flips the image horizontally with a probability of 0.5
    # transforms.RandomRotation(degrees=15),   # Rotates the image by up to +/- 15 degrees
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])


class ParksDataset(Dataset):
    def __init__(self, image_paths, mask_paths, patch_size=256, transform=None):
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.patch_size = patch_size
        self.transform = transform

        self.images = [tiff.imread(image_path) for image_path in image_paths]
        self.masks = [tiff.imread(mask_path) for mask_path in mask_paths]

        for img, mask in zip(self.images, self.masks):
            assert img.shape[:2] == mask.shape[:2], "Image and mask must have the same spatial dimensions."

        self.patches = []
        for idx, (image, mask) in enumerate(zip(self.images, self.masks)):
            height, width = image.shape[:2]
            patches_x = width // patch_size
            patches_y = height // patch_size
            self.patches.extend([(idx, x, y) for y in range(patches_y) for x in range(patches_x)])

    def __len__(self):
        return len(self.patches)

    def __getitem__(self, idx):
        img_idx, x_idx, y_idx = self.patches[idx]
        x_start = x_idx * self.patch_size
        y_start = y_idx * self.patch_size

        image_patch = self.images[img_idx][y_start:y_start+self.patch_size, x_start:x_start+self.patch_size]
        mask_patch = self.masks[img_idx][y_start:y_start+self.patch_size, x_start:x_start+self.patch_size]

        # Preprocess image_patch
        image_patch = np.squeeze(image_patch)  # Remove singleton dimensions
        if image_patch.dtype != np.uint8:
            image_patch = (image_patch * 255).astype(np.uint8)  # Convert to uint8 if necessary

        # Ensure shape is compatible with PIL
        if len(image_patch.shape) == 2 or (len(image_patch.shape) == 3 and image_patch.shape[2] == 3):
            image_patch = Image.fromarray(image_patch)
        else:
            raise ValueError(f"Unexpected image shape: {image_patch.shape}")

        # Apply transformations if any
        if self.transform:
            image_patch = self.transform(image_patch)

        # Transform mask into one-hot encoding
        # mask_one_hot = np.zeros((2, mask_patch.shape[0], mask_patch.shape[1]), dtype=np.float32)
        # mask_one_hot[1, mask_patch == 1] = 1  # Class 1 for "black" pixels
        # mask_one_hot[0, mask_patch == 0] = 1  # Class 0 for "white" pixels

        return image_patch, mask_patch

```

# Data Loading
I found that a default trian-val-test split at 70%, 15%, and 15% worked best.

```python
dataset = ParksDataset(config['image_paths'], config['mask_paths'], patch_size=config['patch_size'], transform=transform)

# Split the dataset into Train/Validation/Test
total_samples = len(dataset)
train_size = int(0.7 * total_samples)  # 70% for training
val_size = int(0.15 * total_samples)   # 15% for validation
test_size = total_samples - train_size - val_size  # Remaining for testing

train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

# Create DataLoader
train_loader = DataLoader(train_dataset, batch_size=config['batch_size'], shuffle=True, num_workers=4)
val_loader = DataLoader(val_dataset, batch_size=config['batch_size'], shuffle=False, num_workers=4)
test_loader = DataLoader(test_dataset, batch_size=config['batch_size'], shuffle=False, num_workers=4)

# Check DataLoader
for batch_idx, (inputs, targets) in enumerate(train_loader):
    print(f"Batch {batch_idx + 1}")
    print(f"Inputs shape: {inputs.shape}")
    print(f"Targets shape: {targets.shape}")
    print(f"Targets: {targets}")
    break
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

def calculate_class_distribution(dataloader):
    """
    Calculates the pixel distribution of classes in the dataset.
    Args:
        dataloader: DataLoader to fetch the dataset (e.g., train, val, or test loader).
    Returns:
        A dictionary with class labels as keys and pixel counts as values.
    """
    class_counts = Counter()

    for _, masks in dataloader:
        masks = masks.numpy()  # Convert to numpy array
        # masks = masks[:, 1, :, :] # convert from one hot
        unique, counts = np.unique(masks, return_counts=True)
        class_counts.update(dict(zip(unique, counts)))

    return class_counts

def visualize_class_distribution(class_counts):
    """
    Visualizes the class distribution as a bar chart.
    Args:
        class_counts: Dictionary with class labels as keys and pixel counts as values.
    """
    classes = list(class_counts.keys())
    counts = list(class_counts.values())

    plt.figure(figsize=(8, 5))
    plt.bar(classes, counts, tick_label=[f"Class {c}" for c in classes])
    plt.xlabel("Class Labels")
    plt.ylabel("Pixel Counts")
    plt.title("Class Distribution in Dataset")
    plt.show()

# Example usage
class_counts = calculate_class_distribution(train_loader)
print("Class Distribution:", class_counts)

# Visualize the class distribution
visualize_class_distribution(class_counts)
```

# Unet Implementation


```python
class DecoderBlock(nn.Module):
    """Upscaling then double conv"""

    def __init__(self, conv_in_channels, conv_out_channels, up_in_channels=None, up_out_channels=None):
        super().__init__()
        """
        eg:
        decoder1:
        up_in_channels      : 1024,     up_out_channels     : 512
        conv_in_channels    : 1024,     conv_out_channels   : 512

        decoder5:
        up_in_channels      : 64,       up_out_channels     : 64
        conv_in_channels    : 128,      conv_out_channels   : 64
        """
        if up_in_channels==None:
            up_in_channels=conv_in_channels
        if up_out_channels==None:
            up_out_channels=conv_out_channels

        self.up = nn.ConvTranspose2d(up_in_channels, up_out_channels, kernel_size=2, stride=2)
        self.conv = nn.Sequential(
            nn.Conv2d(conv_in_channels, conv_out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(conv_out_channels),
            nn.ReLU(inplace=True),
            nn.Conv2d(conv_out_channels, conv_out_channels, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(conv_out_channels),
            nn.ReLU(inplace=True)
        )

    # x1-upconv , x2-downconv
    def forward(self, x1, x2):
        x1 = self.up(x1)
        x = torch.cat([x1, x2], dim=1)
        return self.conv(x)

class UnetResnet34(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        resnet34 = models.resnet34(pretrained=True)
        filters = [64, 128, 256, 512]

        self.firstlayer = nn.Sequential(*list(resnet34.children())[:3])
        self.maxpool = list(resnet34.children())[3]
        self.encoder1 = resnet34.layer1
        self.encoder2 = resnet34.layer2
        self.encoder3 = resnet34.layer3
        self.encoder4 = resnet34.layer4

        self.bridge = nn.Sequential(
            nn.Conv2d(filters[3], filters[3]*2, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(filters[3]*2),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2)

        )

        self.decoder1 = DecoderBlock(conv_in_channels=filters[3]*2, conv_out_channels=filters[3])
        self.decoder2 = DecoderBlock(conv_in_channels=filters[3], conv_out_channels=filters[2])
        self.decoder3 = DecoderBlock(conv_in_channels=filters[2], conv_out_channels=filters[1])
        self.decoder4 = DecoderBlock(conv_in_channels=filters[1], conv_out_channels=filters[0])
        self.decoder5 = DecoderBlock(
            conv_in_channels=filters[1], conv_out_channels=filters[0], up_in_channels=filters[0], up_out_channels=filters[0]
        )

        self.lastlayer = nn.Sequential(
            nn.ConvTranspose2d(in_channels=filters[0], out_channels=filters[0], kernel_size=2, stride=2),
            nn.Conv2d(filters[0], num_classes, kernel_size=3, padding=1, bias=False)
        )

    def forward(self, x):
        e1 = self.firstlayer(x)
        maxe1 = self.maxpool(e1)
        e2 = self.encoder1(maxe1)
        e3 = self.encoder2(e2)
        e4 = self.encoder3(e3)
        e5 = self.encoder4(e4)

        c = self.bridge(e5)

        d1 = self.decoder1(c, e5)
        d2 = self.decoder2(d1, e4)
        d3 = self.decoder3(d2, e3)
        d4 = self.decoder4(d3, e2)
        d5 = self.decoder5(d4, e1)

        out = self.lastlayer(d5)
        # out = torch.sigmoid(out) # for binary classification

        return out
```

```python
import torch
import math

def focal_tversky(y_pred, y_true, alpha=0.7, beta=0.3, gamma=0.75, class_weights=None):
    # Ensure y_pred is in the probability space
    y_pred = torch.sigmoid(y_pred)  # Convert logits to probabilities if they are not already

    # Calculate True Positives, False Positives, and False Negatives
    TP = torch.sum(y_pred * y_true, dim=(1, 2))  # Sum over height and width dimensions
    FP = torch.sum(y_pred * (1 - y_true), dim=(1, 2))
    FN = torch.sum((1 - y_pred) * y_true, dim=(1, 2))

    # Calculate the Tversky index
    smooth = 1e-6  # A small number to avoid division by zero
    Tversky_index = (TP + smooth) / (TP + alpha * FP + beta * FN + smooth)

    # Calculate the focal loss based on the Tversky index
    focal_loss = torch.pow((1 - Tversky_index), gamma)

    # Apply class weights if provided
    if class_weights is not None:
        # Ensure class_weights is a tensor
        class_weights = torch.tensor(class_weights, device=y_pred.device, dtype=y_pred.dtype)
        # Reshape class weights to be broadcastable over the batch and spatial dimensions
        class_weights = class_weights.view(1, -1, 1, 1)
        # Weight the loss for each class
        print(class_weights.shape)
        print(class_weights)
        focal_loss = focal_loss * class_weights

    return torch.mean(focal_loss)  # Return the mean loss over the batch


```

# Training Loop
I used a standard Cross Entropy Loss with an Adam optimizer. I made sure to stop training at convergence of delta_loss < 0.001

```python
import torch

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Instantiate model, loss function, and optimizer
total_pixels = sum(class_counts.values())
weights = [total_pixels / class_counts[c] for c in sorted(class_counts.keys())]
weights = torch.tensor(weights).float().to(device)
print("total_pixels", total_pixels)
print("class_counts", class_counts)
print("weights", weights)
loss_function = nn.CrossEntropyLoss(
    weight=weights
    )
# loss_function = focal_tversky


model = UnetResnet34().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# Dictionaries to store loss and accuracy history
history = {
    "train_loss": [],
    "val_loss": [],
    "train_accuracy": [],
    "val_accuracy": [],
}

best_val_loss = float('inf')  # Initialize best validation loss
best_epoch = 0
best_model_path = "best_model.pth"  # File path to save the best model

# Training loop
delta_loss = 1000
for epoch in range(config['epoch']):
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()  # Clear existing gradients
        outputs = model(inputs)  # Forward pass
        # outputs = outputs[:, 1, :, :]
        # print(outputs.shape)
        # print(labels.shape)
        labels = labels.long()
        # print(labels.dtype)
        # print(outputs.dtype)
        loss = loss_function(outputs, labels,
                            #  class_weights=weights
                             )  # Compute loss
        loss.backward()  # Backpropagate the loss
        optimizer.step()  # Update model weights

        print("Batch Loss:", loss.item())  # Optionally print the loss for the current batch

        running_loss += loss.item()




    train_loss = running_loss / len(train_loader)
    # print(f"train loss: {train_loss}")
    # print(f"running loss: {running_loss}")
    # print(f"len train loader: {len(train_loader)}")
    history["train_loss"].append(train_loss)

    # Validation loop
    model.eval()
    running_val_loss = 0.0

    for val_inputs, val_labels in val_loader:
        val_inputs, val_labels = val_inputs.to(device), val_labels.to(device)
        val_outputs = model(val_inputs)
        # val_outputs = val_outputs[:, 1, :, :]
        val_labels = val_labels.long()
        val_loss = loss_function(val_outputs, val_labels,
                                #  class_weights=weights
                                 )
        running_val_loss += val_loss.item()

        # Compute IoU here if desired

    val_loss = running_val_loss / len(val_loader)
    history["val_loss"].append(val_loss)

    print(f"Epoch {epoch + 1}:")
    print(f"Train Loss: {train_loss:.4f}, Train Accuracy: ")
    print(f"Val Loss: {val_loss:.4f}, Val Accuracy: ")

    # Save the best model based on validation loss
    if val_loss < best_val_loss:
        best_val_loss = val_loss
        best_epoch = epoch
        torch.save(model.state_dict(), best_model_path)
        print(f"Best model saved at epoch {epoch + 1} with validation loss: {val_loss:.4f}")

    delta_loss = train_loss

# Save history to a file (optional)
import json
with open("training_history.json", "w") as f:
    json.dump(history, f)

# Print the best val loss and epoch
print(f"Best Validation Loss: {best_val_loss:.4f}")
print(f"Epoch of Best Validation Loss: {best_epoch + 1}")
```

```python
import seaborn as sns

def evaluate_model_with_confusion_matrix_and_accuracy(model, data_loader, num_classes=2):
    model.eval()
    all_true = torch.tensor([]).to(device)
    all_preds = torch.tensor([]).to(device)

    with torch.no_grad():
        for inputs, labels in data_loader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            # labels = labels[:, 1, :, :]
            outputs = model(inputs)
            outputs = outputs[:, 1, :, :]
            predictions = (outputs > 0.5).long()  # Convert probabilities to 0 or 1
            # predictions = predictions[:, 1, :, :]
            # print(predictions.shape)

            # Reshape labels and predictions to one dimension
            all_true = torch.cat((all_true, labels))
            all_preds = torch.cat((all_preds, predictions))

    # Compute confusion matrix for each class
    cm = torch.zeros(num_classes, num_classes)
    for i in range(num_classes):
        for j in range(num_classes):
            print(all_preds.shape)
            print(all_true.shape)
            cm[i, j] = ((all_preds == i) & (all_true == j)).sum().item()

    # Calculate accuracy
    accuracy = (all_preds == all_true).sum().item() / all_true.numel()

    return cm.cpu().numpy(), accuracy

# Compute confusion matrix and accuracy on test set
cm, accuracy = evaluate_model_with_confusion_matrix_and_accuracy(model, test_loader)

# Print accuracy
print(f"Accuracy: {accuracy * 100:.2f}%")

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='.1f', cmap='Blues', xticklabels=['Background', 'Object'], yticklabels=['Background', 'Object'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import numpy as np

def calculate_multiclass_precision_recall_f1(cm):
    num_classes = cm.shape[0]
    precision = np.zeros(num_classes)
    recall = np.zeros(num_classes)
    f1_scores = np.zeros(num_classes)

    for i in range(num_classes):
        TP = cm[i, i]
        FP = cm[:, i].sum() - TP
        FN = cm[i, :].sum() - TP

        precision[i] = TP / (TP + FP) if (TP + FP) > 0 else 0
        recall[i] = TP / (TP + FN) if (TP + FN) > 0 else 0
        if (precision[i] + recall[i]) > 0:
            f1_scores[i] = 2 * (precision[i] * recall[i]) / (precision[i] + recall[i])
        else:
            f1_scores[i] = 0

    # Macro average
    macro_precision = np.mean(precision)
    macro_recall = np.mean(recall)
    macro_f1_score = np.mean(f1_scores)

    return macro_precision, macro_recall, macro_f1_score

# Use your confusion matrix 'cm'
macro_precision, macro_recall, macro_f1_score = calculate_multiclass_precision_recall_f1(cm)

print(f"Macro Precision: {macro_precision:.4f}")
print(f"Macro Recall: {macro_recall:.4f}")
print(f"Macro F1 Score: {macro_f1_score:.4f}")
```

```python
import matplotlib.pyplot as plt

def visualize_predictions(model, dataloader, num_images=10):
    """
    Visualizes predictions from the model alongside input images and ground truth masks.
    Args:
        model: Trained PyTorch model.
        dataloader: DataLoader to fetch data.
        num_images: Number of images to visualize.
    """
    model.eval()
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    images, true_masks, predicted_masks = [], [], []

    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)

            # Adjust for grayscale inputs if necessary
            if inputs.shape[1] == 1:
                inputs = inputs.repeat(1, 3, 1, 1)

            outputs = model(inputs)

            # Assuming outputs are logits for each class, apply sigmoid and threshold to get binary mask
            preds = (outputs > 0.5).long()  # Convert probabilities to 0 or 1

            # Move tensors to CPU for visualization
            images.append(inputs.cpu())
            true_masks.append(labels.cpu())
            predicted_masks.append(preds.cpu())


            if len(images) >= num_images:
                break

    # Plotting the results
    for i in range(num_images):
        fig, ax = plt.subplots(1, 3, figsize=(15, 5))

        # Input image
        img = images[0][i].numpy().transpose(1, 2, 0)
        img = (img - img.min()) / (img.max() - img.min())  # Normalize image for display
        ax[0].imshow(img)
        ax[0].set_title("Input Image")
        ax[0].axis("off")

        # Ground truth mask
        # print(true_masks[0][i][0])
        gt_mask = true_masks[0][i]  # Assuming first channel for visual (change if different)
        ax[1].imshow(gt_mask, cmap='gray')
        ax[1].set_title("Ground Truth Mask")
        ax[1].axis("off")

        # Predicted mask
        print(torch.min(predicted_masks[0][i][0]), torch.max(predicted_masks[0][i][0]))
        predicted_masks[0][i][1]
        pred_mask = predicted_masks[0][i][0]  # Extract a single predicted mask
        ax[2].imshow(pred_mask, cmap='gray')  # Display as grayscale
        ax[2].set_title("Predicted Mask")
        ax[2].axis("off")

        plt.show()

# Assuming 'model' and 'test_loader' are already defined
# Call the function to visualize
visualize_predictions(model, test_loader)
```
